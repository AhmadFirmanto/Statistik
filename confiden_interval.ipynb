#confidence interval

```python
sampel = np.array([20, 25, 37, 20, 23, 21, 11, 14, 12, 24])

data_sampel = pd.DataFrame(sampel)
data_sampel
```

* mengambil 10 data untuk di jadikan sampel

```python
data_sampel.describe()
```

* standar deviasi sample = 7.57
* n sampel = 10
* cl = 95%
* rata-rata sampel = 20.70

```python
confiden_interval = stats.norm.interval(0.95, loc=20.70, scale=7.57/sqrt(10))
confiden_interval

#interval 95%
#loc= rata-rata sampel
#scale= rumus confiden interfal = standar deviasi sampel / akar rata-rata sampel
```
