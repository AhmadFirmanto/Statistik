#Library

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import random
```

#Data

```python
data_a = np.array([random.choice(range(20, 50)) for i in range(30)])
data_b = np.array([random.choice(range(10, 40)) for i in range(30)])

data = pd.DataFrame({'A' : data_a, 'B' : data_b})
data
```

##rata-rata

```python
np.mean(data)
```

* rata-rata data_a = 34.16
* rata-rata data_b = 23.10

```python
data.describe()
```

* count(data) = 30
* mean(rata-rata)
  - data_a = 34.16
  - data_b = 23.10
* std(standart deviasi)
  - data_a = 8.42
  - data_b = 8.46
* 25%(kuartil bawah)q1
  - data_a = 28
  - data_b = 15.25
* 50%(median)q2
  - data_a = 34
  - data_b = 21
* 75%(kuartil atas)q3
  - data_a = 43.5
  - data_b = 31.25
